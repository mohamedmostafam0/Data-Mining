```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import KBinsDiscretizer, MinMaxScaler
from sklearn.model_selection import train_test_split, KFold, StratifiedKFold
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
```

```python
from apyori import apriori
import pyfpgrowth
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Mall_Customers.csv')

# Understand dataset
print("Dataset Head:", data.head())
print("Dataset Info:", data.info())
print("Descriptive Statistics:", data.describe())
```

```python
# 1. Data Cleaning and Imputation
# Handle missing values
imputer = SimpleImputer(strategy='mean')
data['Age'] = imputer.fit_transform(data[['Age']])
```

```python
# 2. Outlier Detection
# Detect outliers in the 'Annual Income' and 'Spending Score' features
sns.boxplot(data['Annual Income (k$)'])
plt.title('Outliers in Annual Income')
plt.show()

sns.boxplot(data['Spending Score (1-100)'])
plt.title('Outliers in Spending Score')
plt.show()
```

```python
# 3. Feature Scaling and Discretization
scaler = MinMaxScaler()
data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']] = scaler.fit_transform(data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
```

```python
discretizer = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
data['Income_bins'] = discretizer.fit_transform(data[['Annual Income (k$)']])
```

```python
# 4. Data Splitting Techniques
X = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
kf = KFold(n_splits=4)
stratified_kf = StratifiedKFold(n_splits=4)
```

```python
# Example split with train-test split
X_train, X_test = train_test_split(X, test_size=0.3, random_state=42)
```

```python
# 5. Clustering Models
# K-Means Clustering
kmeans = KMeans(n_clusters=4)
kmeans_labels = kmeans.fit_predict(X)
data['KMeans_Labels'] = kmeans_labels
```

```python
# DBSCAN Clustering
dbscan = DBSCAN(eps=0.3, min_samples=4)
dbscan_labels = dbscan.fit_predict(X)
data['DBSCAN_Labels'] = dbscan_labels
```

```python
# Agglomerative Clustering
agg_clust = AgglomerativeClustering(n_clusters=4)
agg_labels = agg_clust.fit_predict(X)
data['Agg_Clust_Labels'] = agg_labels
```

```python
# 6. Association Rule Mining (for discrete features)
transactions = data[['Income_bins', 'Spending Score (1-100)']].values.tolist()
rules = apriori(transactions, min_support=0.2, min_confidence=0.5)
association_rules = list(rules)
```

```python
# Visualization
# Correlation Matrix
numeric_data = data.select_dtypes(include=[np.number])
corr = numeric_data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# KMeans Clustering Results
plt.scatter(numeric_data['Annual Income (k$)'], numeric_data['Spending Score (1-100)'], c=numeric_data['KMeans_Labels'], cmap='viridis')
plt.title('KMeans Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
# DBSCAN Clustering Results
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], c=data['DBSCAN_Labels'], cmap='plasma')
plt.title('DBSCAN Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
# Agglomerative Clustering Results
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], c=data['Agg_Clust_Labels'], cmap='inferno')
plt.title('Agglomerative Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
# 7. Association Rules Result
print("Association Rules:")
for rule in association_rules:
    print(rule)
```
